### 2. TensorFlow'u İçe Aktarma

TensorFlow'u projede kullanmak için içe aktarmanız gerekir:

```python
import tensorflow as tf
```

### 3. Veri Hazırlama

Modelinizi eğitmek için verilerinizi hazırlamanız gerekir. Aşağıda, NumPy kullanarak rastgele veriler oluşturma örneği bulunmaktadır:

```python
import numpy as np

# Eğitim ve test verileri oluşturma
x_train = np.random.rand(1000, 32)
y_train = np.random.randint(2, size=(1000, 1))

x_test = np.random.rand(200, 32)
y_test = np.random.randint(2, size=(200, 1))
```

### 4. Model Oluşturma

TensorFlow'da bir model oluşturmak için `Sequential` API veya `Functional` API kullanılabilir. İşte basit bir Sequential model örneği:

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(32,)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

### 5. Modeli Derleme

Modeli derlemek için `compile` metodunu kullanabilirsiniz. Bu aşamada, kayıp fonksiyonunu, optimizasyon algoritmasını ve değerlendirme metriğini belirtmeniz gerekir:

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### 6. Modeli Eğitme

Modeli eğitmek için `fit` metodunu kullanabilirsiniz:

```python
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test))
```

### 7. Modeli Değerlendirme

Eğitilmiş modelinizi değerlendirmek için `evaluate` metodunu kullanabilirsiniz:

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test loss: {loss}')
print(f'Test accuracy: {accuracy}')
```

### 8. Tahmin Yapma

Modelinizle yeni veriler üzerinde tahmin yapmak için `predict` metodunu kullanabilirsiniz:

```python
predictions = model.predict(x_test)
print(predictions)
```

### 9. Modeli Kaydetme ve Yükleme

Eğitilmiş modelinizi kaydetmek için `save` metodunu kullanabilirsiniz:

```python
model.save('my_model.h5')
```

Kaydedilmiş bir modeli yüklemek için `load_model` fonksiyonunu kullanabilirsiniz:

```python
from tensorflow.keras.models import load_model

model = load_model('my_model.h5')
```
